```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

# Data dictionary
data = {
    "SIZE(sqft)": [151, 234, 332, 512, 211, 551, 377],
    "COST(x1000$)": [4, 6, 11, 14, 5, 15, 11.5]
}

# Create DataFrame
df = pd.DataFrame(data)

# Prepare training data
x_train = np.array(df['SIZE(sqft)'])
y_train = np.array(df['COST(x1000$)'])

# Scatter plot
plt.scatter(x_train, y_train, marker='x')
plt.xlabel('SIZE(sqft)')
plt.ylabel('COST(x1000$)')
plt.show()

def gradient(x, y, w, b):
    dj_dw = 0
    dj_db = 0
    m = x.shape[0]
    
    # Vectorized calculation
    for i in range(m):
        fb = (w * x[i]) + b  
        error = fb - y[i]  
        
        # Update gradients
        dj_dw += error * x[i]
        dj_db += error

    # Average gradients
    dj_dw /= m
    dj_db /= m
    return dj_dw, dj_db

def gradient_descent(x, y, w_init, b_init, alpha):
    w = w_init
    b = b_init
    num_iterations = 1000
    
    for i in range(num_iterations):
        dj_dw, dj_db = gradient(x, y, w, b)
        w -= alpha * dj_dw  
        b -= alpha * dj_db  
        
    return w, b

# Initial parameters
w_init = 0
b_init = 0
alpha = 1.0e-5

# Run gradient descent
w, b = gradient_descent(x_train, y_train, w_init, b_init, alpha)

# Output results
w, b

#Regression
plt.scatter(x_train, y_train, marker='x')
a=np.linspace(0,600,7)
c=w*a+b
plt.plot(a,c)
plt.show()

#Estimator
for i in range(7):
size=int(input("Enter Size of House"))
print(f"Estimated Price of House is {((w*size+b)*1000):0.0f}$")
```
