```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
height=np.array([160,180,170,180,160,151,170,190,167])
weight=np.array([54,97,60,72,80,50,90,77,54])
```

```python
h=([])
for i in range(9):
    if height[i]>=180 or weight[i]>=80:
        h.append(1)
    else:
        h.append(0)

huge=np.array(h)
```

```python
plt.scatter(height,weight,marker='x',color='b')

plt.xlabel='Weight'
plt.ylabel='Height'
plt.show()
```

```python
def sigmoid(z):
    fb=1/(1+np.exp(-z))
    return fb
```

```python
def gradient(x,y,huge,w,b):
    m,n=9,2
    dj_dw=np.zeros(n)
    dj_db=0

    for i in range(m):
        fb=sigmoid(w[0]*x[i]+w[1]*y[i]+b)
        err=fb-huge[i]
        dj_dw[0]=dj_dw[0]+err*x[i]
        dj_dw[1]=dj_dw[1]+err*y[i]
        dj_db=dj_db+err

    return dj_dw,dj_db
```

```python
def gradient_descent(x,y,huge,w,b,aplha):
    dj_dw,dj_db=gradient(x,y,huge,w,b)

    for i in range(1000):
        w=w-aplha*dj_dw
        b=b-alpha*dj_db

    return w,b
        
```

```python
w=[0,0]
b=0
alpha=0.1

w,b=gradient_descent(height,weight,huge,w,b,alpha)
print(w,b)
```
